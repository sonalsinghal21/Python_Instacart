# 7. You’ll also need to explore whether there’s a connection between age and spending power (income). To visualize this relationship, create a scatterplot using the sns.scatterplot() function.

```python
# scatterplot of age and income
scatter_age_income = sns.scatterplot(x = 'Age', y = 'Income',data = df_ords_prods_cust)
```

This scatterplot shows that most people up to the age of 40 earn under 200k and the max earned under 40 years old is 400k. Then from age 40+ more people are in the 200k-300k range and the max goes up to 600k therefore the spending power is higher in those aged over 40.

```python
#7 scatterplot of age_bracket and income
scatter_age_bracket_income = sns.scatterplot(x = 'Age_bracket', y = 'Income',data = df_ords_prods_cust)
```

Observation: The Majority of individuals across all age groups have income below 200k. However , around the age of 40 and above, there is an increase in the proportion of the population earning up to 300k. No individuals below the age of 40 have incomes exceeding 400k, but a notable portion of these aged 40 and above earn up to 600k. The main observation is that the income tends to show an upward trend around the age of 40, which aligns with the life stage of individuals. While there are some jumps in income from 400k to 600k or from 300k, the general income change occurs around this age range.
